```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as scipy
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator


%matplotlib inline
```

```python
# import data
df_raw = pd.read_csv('wine-data.csv', names=['id','country','description','designation','points','price',
                                   'province','region_1','region_2','taster_name','taster_twitter_handle',
                                   'title','variety','winery'])


# clean data take out useless columns 
df_raw = df_raw.drop(columns=['taster_name', 'taster_twitter_handle'])

# extract vintage from title
df_raw['year'] = df_raw['title'].str.extract('(\d{4})')

# drop any entries without price or points
df_raw = df_raw.dropna(subset=['points', 'price'])

# create new "value" feature
df_raw['value'] = df_raw['points'] / df_raw['price']

df_raw.head()
```

```python
# scater plot of price vs points
plt.scatter(df_raw['price'], df_raw['points'])
plt.xlabel("Price")
plt.ylabel("Points")
plt.show()
```

```python
# histogram of "value"
plt.hist(df_raw['value'])
```

```python
# code based off of kernel from kaggle
# https://inclass.kaggle.com/scrps93/wine-pricier-is-better/versions

df_wc = df_raw

# break up data into quartiles
score_quart4 = df_wc[df_wc.value >= df_wc.value.quantile(.75)].reset_index(drop=True)
score_quart4 = score_quart4.dropna().reset_index(drop=True)

score_quart3 = df_wc[(df_wc.value < df_wc.value.quantile(.75)) & (df_wc.value >= df_wc.value.quantile(.50))].reset_index(drop=True)
score_quart3 = score_quart3.dropna().reset_index(drop=True)

score_quart2 = df_wc[(df_wc.value < df_wc.value.quantile(.50)) & (df_wc.value >= df_wc.value.quantile(.25))].reset_index(drop=True)
score_quart2 = score_quart2.dropna().reset_index(drop=True)

score_quart1 = df_wc[df_wc.value < df_wc.value.quantile(.25)].reset_index(drop=True)
score_quart1 = score_quart1.dropna().reset_index(drop=True)

qt1_txt = str(score_quart1.description)
qt2_txt = str(score_quart2.description)
qt3_txt = str(score_quart3.description)
qt4_txt = str(score_quart4.description)

# set stop words
stopwords = set(STOPWORDS)
stopwords.update(["drink", "now", "wine", "flavor", "flavors", "aroma", "aromas", "blend"])


# create word clouds
plt.figure(figsize=(15,15))

plt.subplot(2, 2, 1)
plt.title('Q1 by Value (lowest-scoring)', fontsize=15)
wordcloud = WordCloud(stopwords=stopwords, background_color='white').generate(qt1_txt)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.subplot(2, 2, 2)
plt.title('Q2 by Value', fontsize=15)
wordcloud = WordCloud(stopwords=stopwords, background_color='white').generate(qt2_txt)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.subplot(2, 2, 3)
plt.title('Q3 by Value', fontsize=15)
wordcloud = WordCloud(stopwords=stopwords, background_color='white').generate(qt3_txt)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.subplot(2, 2, 4)
plt.title('Q4 by Value (highest-value)', fontsize=15)
wordcloud = WordCloud(stopwords=stopwords, background_color='white').generate(qt4_txt)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python

df = df_raw

df = df.drop(columns=['id', 'description', 'region_2', 'title', 'price'])

# drop entries with missing data
df = df[df['country'].notnull()]
df = df[df['designation'].notnull()]
df = df[df['points'].notnull()]
df = df[df['region_1'].notnull()]
df = df[df['variety'].notnull()]
df = df[df['winery'].notnull()]
df = df[df['year'].notnull()]


# This step is to change the data into numeric attributes to be able to make models of the date
le1 = preprocessing.LabelEncoder()
le1.fit(df['country'])
df['country'] = le1.transform(df['country'])
le2 = preprocessing.LabelEncoder()
le2.fit(df['designation'])
df['designation'] = le2.transform(df['designation'])
le3 = preprocessing.LabelEncoder()
le3.fit(df['points'])
df['points'] = le3.transform(df['points'])
le5 = preprocessing.LabelEncoder()
le5.fit(df['province'])
df['province'] = le5.transform(df['province'])
le5 = preprocessing.LabelEncoder()
le5.fit(df['region_1'])
df['region_1'] = le5.transform(df['region_1'])
le6 = preprocessing.LabelEncoder()
le6.fit(df['variety'])
df['variety'] = le6.transform(df['variety'])
le7 = preprocessing.LabelEncoder()
le7.fit(df['winery'])
df['winery'] = le7.transform(df['winery'])
le8 = preprocessing.LabelEncoder()
le8.fit(df['year'])
df['year'] = le8.transform(df['year'])

df_value = df['value']
category = pd.qcut(df_value,4)
category = category.to_frame()
df['value_buckets'] = category.astype('str')

le9 = preprocessing.LabelEncoder()
le9.fit(df['winery'])
df['variety'] = le9.transform(df['winery'])
le10 = preprocessing.LabelEncoder()
le10.fit(df['winery'])
df['year'] = le10.transform(df['year'])
```

```python
df.head()
```

```python
# select features to use
features = ['country','designation',
                                   'province','region_1',
                                   'variety','winery','year']
X = df[features]
y = df['value_buckets']

# split data into test and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)

# create and test model (entropy)
model = tree.DecisionTreeClassifier(criterion='entropy')
model.fit(X_train, y_train)

y_predict = model.predict(X_test)

accuracy_score(y_test,y_predict)
```

```python
# create and test model (gini)
model = tree.DecisionTreeClassifier(criterion='gini')

model.fit(X_train, y_train)

y_predict = model.predict(X_test)

accuracy_score(y_test,y_predict)
```
